```python
import pandas as pd
import ast
import numpy as np
```

```python
df=pd.read_csv('Data/movies_metadata.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['adult','homepage','imdb_id','overview','poster_path','video'], inplace=True)
```

```python
df[['belongs_to_collection','genres','production_companies','production_countries','spoken_languages']].isna().sum()
```

```python
def stringified(x):
    if isinstance(x, str) == True:
        lst = ast.literal_eval(x)
        return lst
    else:
        return np.nan
```

```python
df[['belongs_to_collection','genres','production_companies','production_countries','spoken_languages']]=df[['belongs_to_collection','genres','production_companies','production_countries','spoken_languages']].applymap(lambda x: stringified(x))
```

```python
df.info()
```

```python
df.head()
```

```python
def name(x):
    if isinstance(x,dict)==True:
        getx=x["name"]
        return getx
    else:
        return np.nan
```

```python
df['belongs_to_collection']=df['belongs_to_collection'].apply(lambda x: name(x))
```

```python
df['genres'][0]
```

```python
cols=['genres','production_companies','production_countries','spoken_languages']
```

```python
df_exploded=df
```

```python
for i in cols:
    df_exploded=df_exploded.explode(i)
```

```python
df_exploded['genres']=df_exploded['genres'].apply(lambda x: name(x))
df_exploded['production_companies']=df_exploded['production_companies'].apply(lambda x: name(x))
df_exploded['production_countries']=df_exploded['production_countries'].apply(lambda x: name(x))
df_exploded['spoken_languages']=df_exploded['spoken_languages'].apply(lambda x: name(x))
```

```python
df_exploded.drop(columns=['tagline'],inplace=True)
```

```python
df_exploded.drop(columns=['original_title'],inplace=True)
```

```python
df_exploded.reset_index(drop=True, inplace=True)
```
